### Exercise 3: Training a Classifier

Lets' repeat what we did manually in the previous lab using scikit-learn. We'll use a LDA classifier, which we can instanciate as follows:

```python
import sklearn.discriminant_analysis as DA
Fisher=DA.LinearDiscriminantAnalysis()
```

As discussed in the lecture, to properly formulate our problem, we'll have to:

* Define the inputs (X) vs outputs (Y)
* Designate training vs testing samples (in order to get a unbias assessment of the performance of Machine Learning algorithms)

for example, here we'll take use 4M events for training and the remainder for testing.

```python
N_Train=400_000

Train_Sample=df[:N_Train] 
Test_Sample=df[N_Train:]

X_Train=Train_Sample[VarNames[1:]]
y_Train=Train_Sample["signal"]

X_Test=Test_Sample[VarNames[1:]]
y_Test=Test_Sample["signal"]

Test_sig=Test_Sample[Test_Sample.signal==1]
Test_bkg=Test_Sample[Test_Sample.signal==0]

```

```python
print(Test_sig.shape)
print(Test_bkg.shape)

```

We can train the classifier as follow:

```python
Fisher.fit(X_Train,y_Train)
```

We can plot the output, comparing signal and background:

```python
plt.figure()
plt.hist(Fisher.decision_function(Test_sig[VarNames[1:]]),bins=100,histtype="step", color="blue", label="signal",stacked=True)
plt.hist(Fisher.decision_function(Test_bkg[VarNames[1:]]),bins=100,histtype="step", color="red", label="background",stacked=True)
plt.legend(loc='upper right')
plt.show()
```

#### Part a

Compare ROC curves computed on the test versus training samples, in a single plot. Do you see a bias?

```python
from sklearn.metrics import roc_curve, auc

#ROC for test
fpr_test, tpr_test, _ = roc_curve(y_Test, Fisher.decision_function(X_Test))
roc_auc_test = auc(fpr_test, tpr_test)

#ROC for train
fpr_train, tpr_train, _ = roc_curve(y_Train, Fisher.decision_function(X_Train))
roc_auc_train = auc(fpr_train, tpr_train)

plt.plot(fpr_train, tpr_train,color='blue',label='Train ROC curve (area = %0.2f)' % roc_auc_train)
plt.plot(fpr_test, tpr_test, color='darkorange',label='Test ROC curve (area = %0.2f)' % roc_auc_test)
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Test Versus Train ROC Curves")

plt.show()
```

### There seems to be no evidence of bias

#### Part b

Train the Fisher performance of using the raw, features, and raw+features as input. Compare the performance one a single plot. 

```python
#Training Fisher using raw (low level) variables

def train_fisher(train_X, train_y):
    Fisher=DA.LinearDiscriminantAnalysis()
    Fisher.fit(train_X,train_y)
    
    #ROC
    fpr, tpr, _ = roc_curve(train_y, Fisher.decision_function(train_X))
    roc_auc = auc(fpr, tpr)
    return fpr, tpr, roc_auc

#Raw
X_Train_raw=Train_Sample[VarNames[1:9]]
y_Train_raw=Train_Sample["signal"]

#Feature
X_Train_feat=Train_Sample[VarNames[9:]]
y_Train_feat=Train_Sample["signal"]

#ROC Curves
fpr, tpr, roc_auc=train_fisher(X_Train,y_Train)
fpr_raw, tpr_raw, roc_auc_raw=train_fisher(X_Train_raw, y_Train_raw )
fpr_feat, tpr_feat, roc_auc_feat=train_fisher(X_Train_feat, y_Train_feat )

plt.plot(fpr, tpr,color='red',label='Raw + Feat ROC curve (area = %0.2f)' % roc_auc, linestyle='-')
plt.plot(fpr_raw, tpr_raw,color='blue',label='Raw ROC curve (area = %0.2f)' % roc_auc_raw, linestyle='--')
plt.plot(fpr_feat, tpr_feat, color='darkorange',label='Feature ROC curve (area = %0.2f)' % roc_auc_feat)
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC Curves")

plt.show()
```
