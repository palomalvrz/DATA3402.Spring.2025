1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
#Creating a matrix class
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values:
            #Initialize from a list of lists of values
            row_len=set(map(len, values))
    
            #If more than one unique value in the set, raise error
            if len(row_len)>1:
                raise ValueError ("Inconsistent row length")
                        
            self.n=len(values) #Rows 
            self.m=len(values[0] if values else 0) #Columns
            self.values=values 

        elif n is not None and m is not None:
            #Initalize with a list of lists of zeros

            self.n=n #Rows
            self.m=m #Columns
            self.values= [[0]*m for _ in range(n)] 

        else:
            raise ValueError("Matrix cannot be initialized")

    #Indexing an element from the matrix
    def __getitem__(self, index):
        if isinstance (index, tuple):
            i, j = index
            return self.values[i][j] #Return tuple with value
        else:
            return self.values[index] #Retun whole row

    #Assigning values to the matrix
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.values[i][j] = value
        else:
            raise ValueError("Invalid index format")

    #Print
    def __repr__(self):
        return "\n".join(str(row) for row in self.values)
```

2. Add the following methods:

```python
#Creating a matrix class
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values:
            #Initialize from a list of lists of values
            row_len=set(map(len, values))
    
            #If more than one unique value in the set, raise error
            if len(row_len)>1:
                raise ValueError ("Inconsistent row length")
                        
            self.n=len(values) #Rows 
            self.m=len(values[0] if values else 0) #Columns
            self.values=values 

        elif n is not None and m is not None:
            #Initalize with a list of lists of zeros

            self.n=n #Rows
            self.m=m #Columns
            self.values= [[0]*m for _ in range(n)] 

        else:
            raise ValueError("Matrix cannot be initialized")

    #Indexing an element from the matrix
    def __getitem__(self, index):
        if isinstance (index, tuple):
            i, j = index
            return self.values[i][j] #Return tuple with value
        else:
            return self.values[index] #Retun whole row

    #Assigning values to the matrix
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.values[i][j] = value
        else:
            raise ValueError("Invalid index format")

    #Print
    def __repr__(self):
        return "\n".join(str(row) for row in self.values)

    #Get dimensions of matrix
    def shape(self):
        return (self.n, self.m)

    #Transpose-switch rows with its columns
    def transpose(self):
        new_matrix=[[0]*self.n for _ in range(self.m)]
    
        for i in range(self.n):
            for j in range(self.m):
                new_matrix[j][i]=self.values[i][j]
        
        #Return matrix instance with transposed matrix values
        return Matrix(values=new_matrix) 

    def row(self,n):
        return Matrix(values=[self.values[n]])

    def column(self,n):
        return Matrix(values=[[row[n]] for row in self.values])

    #Return list of lists
    def to_list(self):
        return self.values 

    def block(self, n_0, n_1, m_0, m_1):
        
        #Extract the values in smaller matrix
        new_values = [row[m_0:m_1] for row in self.values[n_0:n_1]]
        
        #Return smaller matrix as a new Matrix object
        return Matrix(values=new_values)


        
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python

def constant(n,m,c):
    if isinstance (c, int):
        c=float(c)
    return Matrix(values=[[c] * m for _ in range(n)])

def zeros(n,m):
    return Matrix(values=[[float(0)] * m for _ in range(n)])

def ones(n,m):
    return Matrix(values=[[float(1)] * m for _ in range(n)])

def eye(n):
    return Matrix(values=[[0] * i + [1] + [0] * (n - i - 1) for i in range(n)])
```

```python
print("Matrix filled with floats:")
print(constant(3,3,3), "\n")

print("Matrix filled with floats of value 0:")
print(zeros(3,3), "\n")

print("Matrix filled with floats of value 1:")
print(ones(3,3), "\n")

print("Identity Matrix")
print(eye(5))
```

```python
#Creating a matrix class
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values:
            #Initialize from a list of lists of values
            row_len=set(map(len, values))
    
            #If more than one unique value in the set, raise error
            if len(row_len)>1:
                raise ValueError ("Inconsistent row length")
                        
            self.n=len(values) #Rows 
            self.m=len(values[0] if values else 0) #Columns
            self.values=values 

        elif n is not None and m is not None:
            #Initalize with a list of lists of zeros

            self.n=n #Rows
            self.m=m #Columns
            self.values= [[0]*m for _ in range(n)] 

        else:
            raise ValueError("Matrix cannot be initialized")

    #Indexing an element from the matrix
    def __getitem__(self, index):
        if isinstance (index, tuple):
            i, j = index
            return self.values[i][j] #Return tuple with value
        else:
            return self.values[index] #Retun whole row

    #Assigning values to the matrix
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.values[i][j] = value
        else:
            raise ValueError("Invalid index format")

    #Print
    def __repr__(self):
        return "\n".join(str(row) for row in self.values)

    #Get dimensions of matrix
    def shape(self):
        return (self.n, self.m)

    #Transpose-switch rows with its columns
    def transpose(self):
        new_matrix=[[0]*self.n for _ in range(self.m)]
    
        for i in range(self.n):
            for j in range(self.m):
                new_matrix[j][i]=self.values[i][j]
        
        #Return matrix instance with transposed matrix values
        return Matrix(values=new_matrix) 

    def row(self,n):
        return Matrix(values=[self.values[n]])

    def column(self,n):
        return Matrix(values=[[row[n]] for row in self.values])

    #Return list of lists
    def to_list(self):
        return self.values 

    def block(self, n_0, n_1, m_0, m_1):
        
        #Extract the values in smaller matrix
        new_values = [row[m_0:m_1] for row in self.values[n_0:n_1]]
        
        #Return smaller matrix as a new Matrix object
        return Matrix(values=new_values)

    def scalarmul(self, c):
        new_values = [[c * value for value in row] for row in self.values]
        return Matrix(values=new_values) 

    #Test if two matrices have the same dimensions
    def same_dimension(self, other):
        return len(self.values) == len(other.values) and len(self.values[0]) == len(other.values[0])

    def add(self, N):
        if not self.same_dimension(N):
            raise ValueError("Dimensions do not match")
            
        new_values = [[self.values[i][j] + N.values[i][j] for j in range(len(self.values[0]))] for i in range(len(self.values))]
        
        return Matrix(values=new_values) 

    def sub(self, N):
        if not self.same_dimension(N):
            raise ValueError("Dimensions do not match")
    
        new_values = [[self.values[i][j] - N.values[i][j] for j in range(len(self.values[0]))] for i in range(len(self.values))]

        return Matrix(values=new_values)


    def mat_mult(self, N):
        if len(self.values[0]) != len(N.values):  
            raise ValueError("Number of columns in M must match number of rows in N")
    
        new_values = [[sum(self.values[i][k] * N.values[k][j] for k in range(len(N.values))) for j in range(len(N.values[0]))] for i in range(len(self.values))]
    
        return Matrix(values=new_values)


    def element_mult(self, N):
        if not self.same_dimension(N):
            raise ValueError("Dimensions do not match")
    
        new_values = [[self.values[i][j] * N.values[i][j] for j in range(len(self.values[0]))] for i in range(len(self.values))]
    
        return Matrix(values=new_values)


    def equals(self, N):
        if not self.same_dimension(N):
            return False
        
        return self.values == N.values
            
```

```python
#Creating a matrix class
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values:
            #Initialize from a list of lists of values
            row_len=set(map(len, values))
    
            #If more than one unique value in the set, raise error
            if len(row_len)>1:
                raise ValueError ("Inconsistent row length")
                        
            self.n=len(values) #Rows 
            self.m=len(values[0] if values else 0) #Columns
            self.values=values 

        elif n is not None and m is not None:
            #Initalize with a list of lists of zeros

            self.n=n #Rows
            self.m=m #Columns
            self.values= [[0]*m for _ in range(n)] 

        else:
            raise ValueError("Matrix cannot be initialized")

    #Indexing an element from the matrix
    def __getitem__(self, index):
        if isinstance (index, tuple):
            i, j = index
            return self.values[i][j] #Return tuple with value
        else:
            return self.values[index] #Retun whole row

    #Assigning values to the matrix
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.values[i][j] = value
        else:
            raise ValueError("Invalid index format")

    #Print
    def __repr__(self):
        return "\n".join(str(row) for row in self.values)

    #Get dimensions of matrix
    def shape(self):
        return (self.n, self.m)

    #Transpose-switch rows with its columns
    def transpose(self):
        new_matrix=[[0]*self.n for _ in range(self.m)]
    
        for i in range(self.n):
            for j in range(self.m):
                new_matrix[j][i]=self.values[i][j]
        
        #Return matrix instance with transposed matrix values
        return Matrix(values=new_matrix) 

    def row(self,n):
        return Matrix(values=[self.values[n]])

    def column(self,n):
        return Matrix(values=[[row[n]] for row in self.values])

    #Return list of lists
    def to_list(self):
        return self.values 

    def block(self, n_0, n_1, m_0, m_1):
        
        #Extract the values in smaller matrix
        new_values = [row[m_0:m_1] for row in self.values[n_0:n_1]]
        
        #Return smaller matrix as a new Matrix object
        return Matrix(values=new_values)

    def scalarmul(self, c):
        new_values = [[c * value for value in row] for row in self.values]
        return Matrix(values=new_values) 

    #Test if two matrices have the same dimensions
    def same_dimension(self, other):
        return len(self.values) == len(other.values) and len(self.values[0]) == len(other.values[0])

    #Overload operators
    def __add__(self, N):
        if not self.same_dimension(N):
            raise ValueError("Dimensions do not match")
            
        new_values = [[self.values[i][j] + N.values[i][j] for j in range(len(self.values[0]))] for i in range(len(self.values))]
        
        return Matrix(values=new_values) 

    def __sub__(self, N):
        if not self.same_dimension(N):
            raise ValueError("Dimensions do not match")
    
        new_values = [[self.values[i][j] - N.values[i][j] for j in range(len(self.values[0]))] for i in range(len(self.values))]

        return Matrix(values=new_values)

    def __mul__(self, N):
        #Scalar multiplication
        if isinstance (N, (int, float)):
            new_values = [[N * value for value in row] for row in self.values]
            return Matrix(values=new_values) 

        #Matrix multiplication
        elif isinstance (N, Matrix):
            if len(self.values[0]) != len(N.values):  
                raise ValueError("Number of columns in M must match number of rows in N")
    
            new_values = [[sum(self.values[i][k] * N.values[k][j] for k in range(len(N.values))) for j in range(len(N.values[0]))] for i in range(len(self.values))]
            return Matrix(values=new_values)


    def element_mult(self, N):
        if not self.same_dimension(N):
            raise ValueError("Dimensions do not match")
    
        new_values = [[self.values[i][j] * N.values[i][j] for j in range(len(self.values[0]))] for i in range(len(self.values))]
    
        return Matrix(values=new_values)


    def __eq__(self, N):
        if not self.same_dimension(N):
            return False
        
        return self.values == N.values

    def __rmul__(self, other):  
        return self.__mul__(other)
            
```
